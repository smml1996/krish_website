```python
import pandas as pd
import json
```

```python
done = ["EAR", "EYE", "FACE", 
        "HEAD", "HEARING", "MIND", 
        "NOSE", "VERTIGO", "VISION", 
        "MOUTH", "TEETH", "THROAT", "EXTERNAL THROAT",
        "STOMACH", "ABDOMEN", "RECTUM", "STOOL", "BLADDER",
        "KIDNEYS", "PROSTATE GLAND", "URETHRA", "URINE",
        "GENITALIA-MALE", "GENITALIA-FEMALE", "LARYNX AND TRACHEA",
        "RESPIRATION", "COUGH"
        ]
```

```python
def get_num_queries_chapter():
    rows = []
    for chapter_name in done:
        f = open(f"data/chapter_{chapter_name}.json", "r")
        j_data = json.loads(f.read())
        f.close()
        num_queries = 0
        for row_ in j_data["data"]:
            if row_["RemediesCount"] > 0:
                num_queries += 1
        rows.append({"chapter": chapter_name, "queries": num_queries})
    
    return pd.DataFrame(rows)
```

```python
get_num_queries_chapter()
```

```python
from pathlib import Path

def get_count_missing_remedies():
    rows = []
    for chapter_name in done:
        f = open(f"data/chapter_{chapter_name}.json", "r")
        j_data = json.loads(f.read())
        f.close()
        symptoms = set()
        for row_ in j_data["data"]:
            if row_["RemediesCount"] > 0:
                assert(row_["SymptomId"] not in symptoms)
                symptoms.add(row_["SymptomId"])
        
        remedies = set()
        
        if Path(f"data/remedies/chapter_{chapter_name}.json").exists():
            f = open(f"data/remedies/chapter_{chapter_name}.json", "r")
            j_data = json.loads(f.read())
            f.close()
            
            for row_ in j_data["data"]:         
                assert(row_["symptom_"]["SymptomId"] not in remedies)
                if row_["RemediesLocal"] > 0:
                    remedies.add(row_["symptom_"]["SymptomId"])
            assert remedies.issubset(symptoms)
        
        rows.append({"chapter": chapter_name, "n_missing": len(symptoms-remedies), "total": len(symptoms), "done": len(remedies)})
    
    return pd.DataFrame(rows)
```

```python
get_count_missing_remedies()
```
